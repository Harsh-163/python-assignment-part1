```python
# -- Task 1: Data Parsing & Profile Cleaning --

# 1. Raw student data provided
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

# 2. Process each student record
for student in raw_students:
    # Clean the name: Remove extra spaces and convert to Title Case
    clean_name = student["name"].strip().title()
    
    # Convert roll number from string to integer
    clean_roll = int(student["roll"])
    
    # Split marks_str and convert each mark to an integer list
    clean_marks = [int(m) for m in student["marks_str"].split(", ")]
    
    # Validation check: Ensure name contains only alphabetical characters
    # Note: We replace spaces with empty strings to check words correctly
    is_valid = clean_name.replace(" ", "").isalpha()
    status_symbol = "✓ Valid name" if is_valid else "✗ Invalid name"
    
    # Store cleaned data in a new dictionary
    cleaned_profile = {
        "name": clean_name,
        "roll": clean_roll,
        "marks": clean_marks
    }
    cleaned_students.append(cleaned_profile)
    
    # 3. Print the Formatted Profile Card
    print("================================")
    print(f"Student : {clean_name}")
    print(f"Roll No : {clean_roll}")
    print(f"Marks   : {clean_marks}")
    print(f"Validation: {status_symbol}")
    print("================================")

# 4. Special requirement for Roll Number 103
print("\n--- Additional Formatting for Roll No 103 ---")
for student in cleaned_students:
    if student["roll"] == 103:
        print(f"ALL CAPS  : {student['name'].upper()}")
        print(f"lowercase : {student['name'].lower()}")
```

Logic Explanation: Data Parsing & Profile Cleaning (Task 1)

The objective of this task was to transform unstructured raw data into a structured format suitable for analysis. Below are the core Python concepts implemented:

1. Data Normalization (String Methods)
Raw data often contains inconsistent formatting. We used the following methods to standardize the name field:
.strip(): Removes unwanted leading and trailing whitespace.
.title(): Ensures only the first letter of each word is capitalized (e.g., ayesha SHARMA → Ayesha Sharma).
.upper() & .lower(): Used specifically for Roll No 103 to demonstrate case manipulation.

2. Type Casting & Parsing
Since the input data came in as strings, we performed type conversion to allow for mathematical operations later:
int(): Converted the roll string into a numerical integer.
.split(", "): Broke the marks_str into individual string elements based on the comma delimiter.

List Comprehension: [int(m) for m in ...] was used to iterate through the split strings and convert each one into an integer simultaneously, creating a list of marks.

3. Data Validation
To ensure data integrity, we checked the validity of names using:
.isalpha(): This method returns True if all characters in the string are alphabetic.
.replace(" ", ""): Since names contain spaces (which are not alphabetic), we temporarily removed spaces to check if the remaining characters were strictly letters.

4. Output Formatting
f-strings (Formatted String Literals): These allow us to embed expressions inside string literals using curly braces {}. This was used to create the "Profile Cards" with a clean, readable layout.

```python
# --- Task 2: Marks Analysis Using Loops & Conditionals ---

# Initial Data
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

print(f"Analysis for Student: {student_name}\n")

# 1. Loop through subjects and marks to assign grades
print("Subject-wise Performance:")
for i in range(len(subjects)):
    score = marks[i]
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"
    
    print(f"- {subjects[i]}: {score} ({grade})")

# 2. Calculate Statistics
total_marks = sum(marks)
average_marks = round((total_marks / len(marks)),2)

# Finding Highest and Lowest
max_val = max(marks)
min_val = min(marks)

max_subject = subjects[marks.index(max_val)]
min_subject = subjects[marks.index(min_val)]

print("\n--- Summary Statistics ---")
print(f"Total Marks    : {total_marks}")
print(f"Average Marks  : {round(average_marks, 2)}")
print(f"Highest Subject: {max_subject} ({max_val})")
print(f"Lowest Subject : {min_subject} ({min_val})")

# 3. While Loop: Marks-Entry System
print("\n--- New Marks Entry System ---")
new_subjects_count = 0

while True:
    sub_input = input("Enter subject name (or type 'done' to finish): ").strip()
    
    if sub_input.lower() == 'done':
        break
    
    marks_input = input(f"Enter marks for {sub_input}: ")
    
    # Error Handling: Check if input is numeric
    if not marks_input.isdigit():
        print("⚠️ Warning: Please enter a valid numeric value.")
        continue
    
    score_val = int(marks_input)
    
    # Error Handling: Check range 0-100
    if score_val < 0 or score_val > 100:
        print("⚠️ Warning: Marks must be between 0 and 100.")
        continue
    
    # Add valid data to lists
    subjects.append(sub_input)
    marks.append(score_val)
    new_subjects_count += 1
    print(f"Added {sub_input} successfully.")

# Final Report after while loop
final_average = sum(marks) / len(marks)
print("\n--- Final Update ---")
print(f"New subjects added: {new_subjects_count}")
print(f"Updated Average Across All Subjects: {round(final_average, 2)}")
```

To prevent the program from crashing due to bad data, we implemented two layers of validation:

```python
# --- Task 3: Class Performance Summary ---

# Initial class data (List of Tuples)
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# Initialize variables for summary statistics
pass_count = 0
fail_count = 0
student_averages = []
topper_name = ""
highest_avg = 0

# Print Table Header
print(f"{'Name':<18} | {'Average':<7} | {'Status'}")
print("-" * 40)

# Loop through class data
for name, marks in class_data:
    # 1. Compute individual average
    avg = round(sum(marks) / len(marks), 2)
    student_averages.append(avg)
    
    # 2. Assign Result Status
    status = "Pass" if avg >= 60 else "Fail"
    
    # Update pass/fail counts
    if status == "Pass":
        pass_count += 1
    else:
        fail_count += 1
        
    # Check for Topper
    if avg > highest_avg:
        highest_avg = avg
        topper_name = name
    
    # 3. Print Formatted Row
    # <18 ensures the name column is left-aligned with a width of 18 characters
    print(f"{name:<18} |  {avg:<7.2f} | {status}")

# 4. Print Summary Report
class_avg = sum(student_averages) / len(student_averages)

print("-" * 40)
print(f"Passed: {pass_count} | Failed: {fail_count}")
print(f"Class Topper: {topper_name} ({highest_avg})")
print(f"Overall Class Average: {round(class_avg, 2)}")
```

Logic Explanation: Class Performance Summary (Task 3)

This task demonstrates the ability to iterate through complex data structures and generate a structured summary report.

1. Handling Tuples and Unpacking
The class_data is stored as a List of Tuples. In the for loop, we used Tuple Unpacking:

This directly assigns the first element of each tuple to name and the second element (the list of marks) to marks, making the code cleaner and more readable.

2. Table Formatting (Alignment)
To create a professional-looking table in the console, we used f-string alignment modifiers:

{avg:<7.2f}: Ensures the average is printed with 2 decimal places and aligned within a 7-character block.

3. Aggregate Logic
Beyond individual processing, the script maintains "running totals" and "state" to calculate class-wide metrics:

Counters: pass_count and fail_count increment based on conditional logic within the loop.

Topper Logic: A "Maximum Seek" algorithm is used by comparing the current student's average against the highest_avg variable initialized outside the loop.

Class Average: By storing each individual average in the student_averages list, we can easily calculate the final class mean using sum() and len() after the loop finishes.

```python
# --- Task 4: String Manipulation Utility ---

# Original essay data
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# 1. Strip whitespace (Base for all subsequent steps)
clean_essay = essay.strip()
print(f"Step 1 (Cleaned): '{clean_essay}'")

# 2. Convert to Title Case
title_essay = clean_essay.title()
print(f"\nStep 2 (Title Case):\n{title_essay}")

# 3. Count occurrences of "python" (Case-insensitive)
# Since the raw essay was lowercase, we check for "python"
python_count = clean_essay.count("python")
print(f"\nStep 3 (Word Count): 'python' appears {python_count} times.")

# 4. Replace "python" with "Python 🐍"
emoji_essay = clean_essay.replace("python", "Python 🐍")
print(f"\nStep 4 (Replacement):\n{emoji_essay}")

# 5. Split into individual sentences
# We split by ". " to separate the sentences
sentence_list = clean_essay.split(". ")
print(f"\nStep 5 (Split List):\n{sentence_list}")

# 6. Print numbered sentences
print("\nStep 6 (Numbered Sentences):")
for i, sentence in enumerate(sentence_list, 1):
    # Ensure each sentence ends with a period
    if not sentence.endswith("."):
        sentence += "."
    
    # Capitalize the first letter of each sentence for better readability
    formatted_sentence = sentence.capitalize()
    print(f"{i}. {formatted_sentence}")
```
